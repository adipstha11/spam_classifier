```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("spam.csv", encoding="latin-1")
```

```python
data.head()
```

```python
data.describe()
```

We would like to just keep the necessary columns now 

```python
data = data[["v1", "v2"]]

data.columns = ['label', 'message']


data.head()
```

```python
data['message'] = data['message'].str.lower().str.replace(r'[^a-z0-9\s]', '', regex=True).str.strip()
data.head()
```

We would like to encode our spam and label SMS so that we could quantify it and do some calculations in it 

```python
data['label'] = data['label'].map({'ham': 0, 'spam' : 1})

X = data['message']
y = data['label']
data.head()
```

```python
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size = 0.2 , random_state = 1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

we are going to vectorize our labels here: 

```python
vectorizer = TfidfVectorizer()

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)


```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# =========================
# Model Comparison Section
# =========================

import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_recall_fscore_support
)

# Here we are defining all the models we want to compare
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Multinomial Naive Bayes": MultinomialNB(),
    "Linear SVM": LinearSVC(),
    "Random Forest": RandomForestClassifier(
        n_estimators=200, random_state=1, n_jobs=-1
    ),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5)
}

# Now we are creating a list where we will save each model's results
results = []

# Here we are training and evaluating each model one by one
for model_name, clf in models.items():
    print("=" * 70)
    print(f"Model: {model_name}")
    print("=" * 70)

    # First we are fitting the model on the same TF-IDF training data
    clf.fit(X_train_vec, y_train)

    # Next we are generating predictions on the test data
    y_pred = clf.predict(X_test_vec)

    # Now we are calculating accuracy and confusion matrix
    acc = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    # Here we are getting precision, recall, and F1 for both classes
    # Then we are taking class 1 values because class 1 = spam
    precision, recall, f1, support = precision_recall_fscore_support(
        y_test, y_pred, labels=[0, 1], average=None, zero_division=0
    )

    spam_precision = precision[1]
    spam_recall = recall[1]
    spam_f1 = f1[1]

    # We are unpacking the confusion matrix so it's easier to read
    # Layout is [[TN, FP], [FN, TP]]
    tn, fp, fn, tp = cm.ravel()

    # Now we are printing all model metrics clearly
    print("Classification Report:")
    print(classification_report(y_test, y_pred, zero_division=0))

    print("Confusion Matrix:")
    print(cm)

    print(f"Accuracy: {acc:.4f}")
    print(f"Spam Precision (class=1): {spam_precision:.4f}")
    print(f"Spam Recall (class=1):    {spam_recall:.4f}")
    print(f"Spam F1-score (class=1):  {spam_f1:.4f}")
    print(f"False Positives (ham -> spam): {fp}")
    print()

    # Here we are storing all values in a dictionary for our comparison table
    results.append({
        "Model": model_name,
        "Accuracy": acc,
        "Spam Precision": spam_precision,
        "Spam Recall": spam_recall,
        "Spam F1": spam_f1,
        "False Positives (ham->spam)": fp,
        "True Negatives": tn,
        "False Negatives": fn,
        "True Positives": tp
    })

# Now we are turning the collected results into a DataFrame
results_df = pd.DataFrame(results)

# Here we are sorting models by overall performance
results_df_sorted = results_df.sort_values(
    by=["Accuracy", "Spam F1", "Spam Recall"],
    ascending=False
).reset_index(drop=True)

print("\n" + "#" * 70)
print("MODEL COMPARISON TABLE")
print("#" * 70)
display(results_df_sorted)

# We are selecting the top row as best overall model after sorting
best_overall = results_df_sorted.iloc[0]

# Now we are finding the model with minimum false positives
# If there is a tie, we are preferring better spam precision and then accuracy
best_min_fp = results_df.sort_values(
    by=["False Positives (ham->spam)", "Spam Precision", "Accuracy"],
    ascending=[True, False, False]
).iloc[0]

print("\n" + "#" * 70)
print("BEST MODEL SUMMARY")
print("#" * 70)

print(
    f"Best overall model: {best_overall['Model']} "
    f"(Accuracy={best_overall['Accuracy']:.4f}, Spam F1={best_overall['Spam F1']:.4f})"
)

print(
    f"Best at minimizing false positives: {best_min_fp['Model']} "
    f"(False Positives={int(best_min_fp['False Positives (ham->spam)'])}, "
    f"Spam Precision={best_min_fp['Spam Precision']:.4f})"
)
```
